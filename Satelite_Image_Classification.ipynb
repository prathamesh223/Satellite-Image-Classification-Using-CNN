```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Define the path to the extracted dataset directory
data_dir = '/content/data'  # Adjust the path as needed

# Define the paths for the train and validation data folders
train_data_folder = '/content/train_dataset'  # Adjust the path as needed
val_data_folder = '/content/val_dataset'  # Adjust the path as needed

# Create train and validation data folders if they don't exist
os.makedirs(train_data_folder, exist_ok=True)
os.makedirs(val_data_folder, exist_ok=True)

# Get the list of class folders in the dataset directory
class_folders = sorted(os.listdir(data_dir))

# Split the dataset into training and validation sets
for class_folder in class_folders:
    class_path = os.path.join(data_dir, class_folder)
    images = os.listdir(class_path)
    train_images, val_images = train_test_split(images, test_size=0.2, random_state=42)

    # Create class folders in train and validation data folders if they don't exist
    os.makedirs(os.path.join(train_data_folder, class_folder), exist_ok=True)
    os.makedirs(os.path.join(val_data_folder, class_folder), exist_ok=True)

    # Move images to the respective train and validation class folders
    for img in train_images:
        src_path = os.path.join(class_path, img)
        dest_path = os.path.join(train_data_folder, class_folder, img)
        shutil.copy(src_path, dest_path)

    for img in val_images:
        src_path = os.path.join(class_path, img)
        dest_path = os.path.join(val_data_folder, class_folder, img)
        shutil.copy(src_path, dest_path)

print("Dataset split and folders created successfully.")
```
